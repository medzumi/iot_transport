```
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import math

#изначальные данные по x, y, v

data = pd.read_csv('dataV4.csv', sep = ',', encoding = 'utf-8')
data_set_speed = np.array(data[['x', 'y', 'v']][:])
#3D рисунок координаты и скорость
fig = plt.figure()
ax = Axes3D(fig)
x = data_set_speed[:,0]
y = data_set_speed[:,1]
z = data_set_speed[:,2]
ax.scatter3D(x, y, z)
plt.show()
```

```
#изначальные данные по x, y, t
data_set_speed = np.array(data[['x', 'y', 'time']][:])
#3D рисунок координаты и скорость
fig = plt.figure()
ax = Axes3D(fig)
x = data_set_speed[:,0]
y = data_set_speed[:,1]
z = data_set_speed[:,2]
ax.scatter3D(x, y, z)
plt.show()
```
